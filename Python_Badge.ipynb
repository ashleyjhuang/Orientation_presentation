#### Project Euler Problem 1: Find the sum of all the multiples of 3 or 5 below 1000

#### Project Euler Problem 2: By considering the terms in the Fibonacci sequence whose values do not exceed four million, find the sum of the even-valued terms.

```python
def fib_es(x):
    a=1
    b=2
    even_sum = 0

    while b <= x:
        if b%2 == 0:
            even_sum = even_sum + b
        a,b = b, a+b
    
    return(even_sum)

fib_es(4000000)
```

```python
### Way 1: takes too long

def factors(x):
    factors_list = [each for each in range(1, x+1) if x % each == 0]

    rev_factors_list = list(reversed(factors_list))

    for rev in rev_factors_list:
        if sympy.isprime(rev) == True:
            break
    return rev

#largest_prime_factors(600851475143)
#largest_prime_factors(13195)
factors(13195)
```

```python
def factors_2(x):
    largest=0

    while x%2==0:
        largest=2
        x = x // 2
    
    for i in range(3, int(math.sqrt(x))+1, 2):
        while x%i == 0: 
            largest= i
            x = x // i
    
    if x > 2:
        largest = x
    
    return largest
```

```python
print(factors_2(13195))
print(factors_2(600851475143))
```

#### Random Distributions and Histograms

```python
# binominal distribution

binom_data = np.random.binomial(10, 0.5, 10)

plt.hist(binom_data, bins=8, color='skyblue', edgecolor='black')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Frequencies Drawn from Binomial Distribution')
plt.show()
```

```python
# poisson distribution

poi_data = np.random.poisson(5, 10)

plt.hist(poi_data, bins=16, color='green', edgecolor='black')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Frequencies Drawn from Poisson Distribution')
plt.show()
```

```python
# chi square distribution

chi_sq_data = np.random.chisquare(2,4)

plt.hist(chi_sq_data, bins=16, color='orange', edgecolor='black')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Frequencies Drawn from Chi-Square Distribution')
plt.show()
```

#### Binomial and Normal Distribution Convergence 

```python
# binomial distribution convergence

binom_data1 = np.random.binomial(10, 0.5, 10)
binom_data2 = np.random.binomial(10, 0.5, 100)
binom_data3 = np.random.binomial(10, 0.5, 1000)
```

```python
plt.hist(binom_data1, bins=16, color='skyblue', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Frequencies from 10 samples in the Binomial Distribution')
plt.show()
```

```python
plt.hist(binom_data2, bins=16, color='skyblue', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Frequencies from 100 samples in the Binomial Distribution')
plt.show()
```

```python
plt.hist(binom_data3, bins=16, color='skyblue', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Frequencies from 1000 samples in the Binomial Distribution')
plt.show()
```

```python
# Normal distribution convergence

norm_data1 = np.random.normal(100, 10, 10)
norm_data2 = np.random.normal(100, 10, 100)
norm_data3 = np.random.normal(100, 10, 1000)
```

```python
plt.hist(norm_data1, bins=16, color='red', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Frequencies from 10 samples in the Normal Distribution')
plt.show()
```

```python
plt.hist(binom_data2, bins=16, color='red', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Frequencies from 10 samples in the Normal Distribution')
plt.show()
```

```python
plt.hist(binom_data3, bins=16, color='red', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Frequencies from 10 samples in the Normal Distribution')
plt.show()
```

##### We can see from both the normal and binomial distributions that as sample size increases, the distribution shape becomes more bell-shaped. 

##### For the binomial distribution, increasing the sample size will result in a more bell-shaped distribution. Additionally, the probability of success (p) in the binomial number generator is 0.5, which would also help give the distribution it's traditional bell-shaped distribution centered in the middle of the graph.

##### For the normal distribution, we see that increasing the sample size also created a more bell-shaped distribution. As more samples are taken, the shape of the histogram reflects the true shape of the normal distribution with the specified arguments applied.
